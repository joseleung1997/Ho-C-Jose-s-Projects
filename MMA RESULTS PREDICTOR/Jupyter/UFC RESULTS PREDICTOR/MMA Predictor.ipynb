```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier
# Ridge classifier uses ridge regression to classify if the fighter will win or lose.

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=10, direction='forward', cv=split)
# Trains the model using random features
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
# Selected values are scaled between 0 and 1 to allow for ridge regression to perform
```

```python
sfs.fit(df[selected_columns], df['target'])
# Pick the 10 best features for predicting target
```

```python
predictors = list(selected_columns[sfs.get_support()])
# get the list of our selected columns out of the sfs
```

```python
predictors
```

```python
# Testing loops we need at least 1 year (52 weeks) to make predictions for half a year at a time (26 weeks steps).
def backtest(data, model, predictors, start=52, step=26):
    all_predictions = []
    
    seasons = sorted(data['date'].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        train = data[data['date'] < season]
        test = data[data['date'] == season]
            # With default inputs, training set will be 1994-1995, test set will be the first 6 month of 1995
        model.fit(train[predictors], train['target'])
            # Take our training data i.e. predictors to figure out which predictor corrolate to target
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
            # Make predictions on the test data set and convert preds' output from a numpy array to pandas series
            # Keep the index of the preds the same as test set index so its easier to combine
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ['actual', 'prediction']
            # axis = 1 treats test and preds as 2 seperate columns, if axis = 0 they'll be combined into 1 column
        all_predictions.append(combined)
    return pd.concat(all_predictions)
        # In the end combine all data frame into 1

'''Essentially, this backtest function splits our data up by half years and uses the first 2 6 months to predict the
next 6 months. however for it to work, we need 2 semi-annuals worth of data for it to work'''
```

```python
def find_fighter_average(fighter):
    rolling = fighter.rolling(4).mean()
    return rolling
# Finds the previous 4 rows (previous 4 matches) and take the average of all the columns
df_rolling = df_rolling.groupby(['R_fighter'], group_keys=False).apply(find_fighter_average)
```

```python
def shift_col(fighter, col_name):
    next_col = fighter[col_name].shift(-1)
    return next_col
# Take the value from the next bout and shift it back 1 to current row

def add_col(df, col_name):
    return df.groupby("R_fighter", group_keys=False).apply(lambda x:shift_col(x, col_name))
# Grouping by fighter name and applying shift function

df = df.dropna()
df["opp_next"] = add_col(df, "B_fighter")
df["date_next"] = add_col(df, "date")
```
